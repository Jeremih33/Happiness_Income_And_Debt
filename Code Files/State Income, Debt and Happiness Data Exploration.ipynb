We will begin importing our necessary libraries. Beautiful soup to webscrape, pandas to work with our dataframes and numpy to work with our number values

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
```

We will begin webscraping several different tables from the web.

1st will be at: https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_income

This is our income data

```python
income_url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_income'
income_html_text = requests.get(income_url).text

income_soup = BeautifulSoup(income_html_text, 'html.parser')

income_table = income_soup.find_all('table')
income_df = pd.read_html(str(income_table))[1]

income_df.columns = ['Rank', 'State', '2019 Income', '2018 Income', '2017 Income', 
                     '2016 Income', '2015 Income', '2014 Income', '2013 Income', 
                     '2012 Income', '2011 Income', '2010 Income', 
                     '2010 to 2019 Income Percent Change']

income_df['State']
```

We will now webscrape personal debt data

This will be at: https://www.businessinsider.com/personal-finance/average-american-debt#average-american-debt-by-state

```python
personal_debt_url = 'https://www.businessinsider.com/personal-finance/average-american-debt'
personal_debt_html_text = requests.get(personal_debt_url).text

personal_debt_soup = BeautifulSoup(personal_debt_html_text, 'html.parser')

personal_debt_table = personal_debt_soup.find_all('table')
personal_debt_df = pd.read_html(str(personal_debt_table))[1]

personal_debt_df

personal_debt_df.columns = ['State', 'Auto Loan Debt', 'Credit Card Debt', 
                            'Mortgage debt', 'Student loan debt', 
                            'Total personal debt']

personal_debt_df['State']
```

Next we will import a csv containing total state debt information, population, cost of living and happiness scores for our states

```python
state_df = pd.read_csv('total_debt_population_happiness.csv')
state_df
```

Now we will join all of our dataframes

```python
state_df = state_df.join(income_df.set_index('State'), on='State')
state_df = state_df.join(personal_debt_df.set_index('State'), on='State')

state_df
```

Next we will clean/examine our data. We will do this in three steps: Identify/handle missing data, cleaning value types and addressing duplicates

```python
print(pd.isnull(state_df).sum())

# state_df = state_df.dropna()

print(state_df.info())

columns = ['2019 Income', '2018 Income', '2017 Income', 
            '2016 Income', '2015 Income', '2014 Income', 
            '2013 Income', '2012 Income', '2011 Income', 
            '2010 Income', 'Auto Loan Debt', 'Credit Card Debt', 
            'Mortgage debt', 'Student loan debt','Total personal debt']

state_df = state_df.drop('Rank', axis=1)

for column in columns:
  state_df[column] = state_df[column].str.replace("$", "")
  state_df[column] = state_df[column].str.replace(",", "")
  state_df[column] = state_df[column].astype(str).astype(int)

percent = '2010 to 2019 Income Percent Change'
state_df[percent] = state_df[percent].str.replace("%", "")
state_df[percent] = state_df[percent].astype(str).astype(float)

print(state_df.info())
```

Here we will check for duplicates in our data

```python
print(state_df.duplicated())
```

Here we will do some basic math to determine state debt per capita in America, find top states in total personal debt and percent change in income.

```python
Total_debt = state_df['Debt'].sum()
Total_population = state_df['Pop'].sum()

per_capita_us = Total_debt / Total_population

print("Total state debt in US: " + str(Total_debt))
print("State debt per capita (US): " + str(per_capita_us))
```

```python
income_increase = state_df.nlargest(n=5, columns=['2010 to 2019 Income Percent Change'])
personal_debt = state_df.nlargest(n=5, columns=['Total personal debt'])

print('The 5 states with the largest increase in income between 2010 and 2019: ')
print(income_increase['State'])

print('\nThe 5 states with the largest amount of personal debt: ')
print(personal_debt['State'])

print('\n')

average_student = state_df['Student loan debt'].mean()
std_dev_student = state_df['Student loan debt'].std()

print('The average student loan debt across states: ' + str(average_student) + 
      "\nThe standard deviation of student loan debt: " + str(std_dev_student))

high_student_loan_states = state_df[state_df['Student loan debt'] > 6000]
print("States with over $6000 in student loan debt on average:")
print(high_student_loan_states['State'])
```

Here we will calculate a column to find where money goes the furthest (on average). We will take 2019 median income and divide it by cost of living index. We will then have a sorted dataframe that prints the states in order of where money goes the furthest to least far.

```python
state_df['Income To COL Ratio'] = state_df['2019 Income'] / state_df['Cost of living index']

sorted_df = state_df.sort_values(by = "Income To COL Ratio", kind = "mergesort", ascending = False)

sorted_df
```

Of course, we know standard deviations are only useful if we have a normal distribution. We will try to find that out now. We will make our first plot with matplotlib, a histogram of student loan debt.

```python
state_df['Student loan debt'].describe()
```

```python
bins = [3000,3500,4000,4500,5000,5500,6000,6500,7000,7500,8000]

plt.hist(state_df['Student loan debt'], bins = bins, color = '#17e3d9')

plt.xticks(bins)

plt.xlabel('Average student loan debt')
plt.ylabel('US Dollars')

plt.title('Distribution of student loan debt')

plt.show()
```

Next we will plot a line graph of the increase in Income from 2010 to 2019 in Minnesota

```python
plt.figure(figsize=(8,5))

# Establishes our x axis
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]

# Establishes our y axis (Minnesota yearly income)
MN = state_df[state_df['State'] == 'Minnesota']
MN_Income = [int(MN['2010 Income']), int(MN['2011 Income']), int(MN['2012 Income']), int(MN['2013 Income']),
             int(MN['2014 Income']), int(MN['2015 Income']),int(MN['2016 Income']), int(MN['2017 Income']),
             int(MN['2018 Income']), int(MN['2019 Income']),]

# Establishes our y axis (New York yearly income)
NY = state_df[state_df['State'] == 'New York']
NY_Income = [int(NY['2010 Income']), int(NY['2011 Income']), int(NY['2012 Income']), int(NY['2013 Income']),
             int(NY['2014 Income']), int(NY['2015 Income']),int(NY['2016 Income']), int(NY['2017 Income']),
             int(NY['2018 Income']), int(NY['2019 Income']),]

# Establishes our y axis (Wisconsin yearly income)
WI = state_df[state_df['State'] == 'Wisconsin']
WI_Income = [int(WI['2010 Income']), int(WI['2011 Income']), int(WI['2012 Income']), int(WI['2013 Income']),
             int(WI['2014 Income']), int(WI['2015 Income']),int(WI['2016 Income']), int(WI['2017 Income']),
             int(WI['2018 Income']), int(WI['2019 Income']),]

# Plot our lines
plt.plot(years, MN_Income, 'b--', label = 'Minnesota')
plt.plot(years, NY_Income, 'r', label = "New York")
plt.plot(years, WI_Income, 'g', label =  "Wisconsin" )
plt.legend()
plt.xticks(years)

# Place our labels
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('Median Income From 2010-2019')

# Show our plot
plt.show()
```

A box and whisker plot of total personal debt would also be interesting:

```python
auto = state_df['Auto Loan Debt']
credit = state_df['Credit Card Debt']
student = state_df['Student loan debt']

labels = ['Auto loan debt', 'Credit card debt', 'Student loan debt']

boxes = plt.boxplot([auto, credit, student], labels = labels, patch_artist=True)

for box in boxes['boxes']:
  box.set(color='#cdedcd', linewidth=2,facecolor='#e0e0e0')

plt.title('Debt type distribution')
plt.ylabel('US dollars')

plt.show()
```
